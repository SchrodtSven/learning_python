# Seabron dataset taxis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

taxis = sns.load_dataset("taxis")
# print(taxis.head())
taxis.total.sum()

info = 'max: {} min: {}'.format(taxis.pickup.max(), taxis.pickup.min())
print(len(taxis))
taxis['total_tip_fare'] = taxis.total - taxis.tip - taxis.fare - taxis.tolls
taxis.head()
```

```python
taxis["date"] = taxis["pickup"].astype("datetime64[ns]").dt.date
taxis_daily = taxis.groupby(["date","payment"], as_index=False).agg(
    total_passengers = ("passengers","sum"),
    total_amount = ("total","sum")
)
taxis_daily.head()
sns.relplot(data=taxis_daily, x="date", y="total_passengers",
            hue="payment", kind="line", height=5, aspect=2);
# sns.displot(data=taxis_daily, x="total_amount", kind="hist", height=5, aspect=1.5, bins=12)
```

```python
with plt.xkcd():
    sns.displot(data=taxis_daily, x="total_amount", kind="hist",
            height=5, aspect=1.5, bins=12);
```
