## Working on seaborn's <q>titanic</q> data set 

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Foo

```python
df = sns.load_dataset('titanic')

#print(df.age.value_counts(dropna=False))
print("Missing values (age):", df.age.isnull().sum())

mdn_age = df.age.median() # calculating median ...
df['age'] = df['age'].fillna(mdn_age) # ... filling it in empty value columns

print("Missing after clearing op:", df.age.isnull().sum())
df.head(10)
```

Let's define some groups (bins) for ages


```python
bins = [13,19,45,60,80,120]

labels = ["Child", "Teen", "Adult", "Middle-Aged", "Senior"] 

df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)

df.age.value_counts()

df.plot()

plt.show()
```
